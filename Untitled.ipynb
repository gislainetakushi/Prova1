```python
#((𝑃 → 𝑄)∧(𝑅 →𝑆)∧(¬𝑄∨¬𝑆)→(¬𝑃∨¬𝑅)

print(f"P| Q | R | S |𝑃 → 𝑄| 𝑅 →𝑆 | (𝑃 → 𝑄)∧(𝑅 →𝑆)|¬𝑄 |¬𝑆 |¬𝑄∨¬𝑆 |(𝑃 → 𝑄)∧(𝑅 →𝑆)∧(¬𝑄∨¬𝑆)| ¬𝑃 |¬𝑅 |¬𝑃∨¬𝑅 | (𝑃 → 𝑄)∧(𝑅 →𝑆)∧(¬𝑄∨¬𝑆)→(¬𝑃∨¬𝑅)")
n=0
for a in range(0,2,1):
    for b in range(0,2,1) :
        for c in range(0,2,1) :
            for d in range(0,2,1) :
                if a ==0:
                    P="V"
                    NP="F"
                else:
                    P="F"
                    NP="V"
                if b==0:
                    Q ="V"
                    NQ ="F"
                else:
                    Q="F"
                    NQ="V"
                if c==0:
                    R="V"
                    NR="F"
                else:
                    R="F"
                    NR="V"
                if d==0:
                    S="V"
                    NS="F"
                else:
                    S="F"
                    NS="V"
                if P== "V" and Q =="F":
                    PQ ="F"
                else:
                    PQ="V"
                if R =="V" and S =="F":
                    RS="F"
                else:
                    RS="V"
                if PQ =="V" and RS=="V":
                    PQRS ="V"
                else:
                    PQRS ="F"
                if NQ =="F" and NS=="F":
                    NQNS ="F"
                else:
                    NQNS ="V"
                if PQRS =="V" and NQNS =="V":
                    PQRSNQNS = "V"
                else:
                    PQRSNQNS = "F"
                if NP == "F" and NR =="F":
                    NPNR ="F"
                else:
                    NPNR="V"
                if PQRSNQNS == "V" and NPNR =="F":
                    print(f"{P}| {Q} | {R} | {S}  |{𝑃Q}    | {𝑅𝑆}     | {𝑃𝑄𝑅𝑆}    |{NQ} |{N𝑆}    |{NQN𝑆} |{𝑃𝑄𝑅𝑆NQN𝑆}| {N𝑃} |{N𝑅} |{N𝑃N𝑅} | F")
                else:
                    
                    print(f"{P}| {Q} | {R} | {S} |{𝑃Q}    | {𝑅𝑆}     | {𝑃𝑄𝑅𝑆}    |{NQ} |{N𝑆}    |{NQN𝑆} |{𝑃𝑄𝑅𝑆NQN𝑆}| {N𝑃} |{N𝑅} |{N𝑃N𝑅} | V")
                    n=n+1
if n ==0:
    print("é uma contradição e não se encaixa no Dilema Destrutivo (D.D.)")
elif n == 16:
    print(" É uma tautologia e representa o Dilema Destrutivo (D.D.)")
else:
    print("É contingencia, portanto não se encaixa no Dilema Destrutivo (D.D.)")
```
